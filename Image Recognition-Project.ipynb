# Dividing the dataset into training and test set

```python
from sklearn.model_selection import train_test_split
```

# Principle compoent analysis  algorithm

```python
from sklearn import decomposition
```

# Assigning a component for analysis

```python
pca = decomposition.PCA(n_components=70,whiten = True,random_state = 1)
```

# Fitting the training set to generate principle components

```python
pca.fit(x_train)
```

# Transforming the principle components

```python
x_train_pca = pca.transform(x_train)
```

```python
x_test_pca = pca.transform(x_test)
```

```python
x_train_pca.shape
```

```python
x_test_pca.shape
```

# Viewing the principle components

```python
eigen=(np.reshape(x[30],(512,512)).astype(np.float64))
```

# Ploting the images one by one as subplots

```python
fig = plt.figure(figsize = (30,30))
for i in range(30):
    ax = fig.add_subplot(5,6,i+1,xticks=[],yticks=[])
    ax.imshow(pca.components_[i].reshape(eigen.shape),cmap = plt.cm.bone)
```

# Support Vector Machine

```python
from sklearn import svm
```

```python
clf = svm.SVC(C=2,gamma = 0.006,kernel='rbf')
clf.fit(x_train_pca,y_train)
```

# Image Prediction

```python
y_pred = clf.predict(x_test_pca)
```

```python
y_pred
```

# Prediction visualization

```python
for i in (np.random.randint(0,4,4)):
    predicted_images=(np.reshape(x_test[i],(512,512)).astype(np.float64))
    plt.title('predicted label: {0}'.format(y_pred[i]))
    plt.imshow(predicted_images,interpolation = 'nearest',cmap = 'gray')
    plt.show
```
